```python
# -*- coding: utf-8 -*-
"""
@author: Morgan
"""

# Depth First Search


graphOfTree = {
    
    '1' : ['2', '5'],
    '2' : ['3', '4'],
    '3' : [],
    '4' : [],
    '5' : ['6', '7'],
    '6' : [],
    '7' : []
    
    }

nodesVisited = set()
def depthFirstSearch(nodesVisited, graphOfTree, node):
    
    if node not in nodesVisited:
        print(node)
        nodesVisited.add(node)
        #print(graphOfTree[node])
        for item in graphOfTree[node]:

            depthFirstSearch(nodesVisited, graphOfTree, item)
            
            
depthFirstSearch(nodesVisited, graphOfTree, '1')
```
