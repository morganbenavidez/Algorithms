```python
# -*- coding: utf-8 -*-
"""
@author: Morgan
"""

# Breadth First Search


graphOfTree = {
    
    
    '1' : ['2', '5'],
    '2' : ['3', '4'],
    '3' : [],
    '4' : [],
    '5' : ['6', '7'],
    '6' : [],
    '7' : []
    
    }


queue = []

def breadthFirstSearch(graphOfTree, node):
    visitedNodes = []
    visitedNodes.append(node)
    queue.append(node)
    
    while queue:
        view = queue.pop(0)
        #print(view, end = ' ')
        
        for item in graphOfTree[view]:
            if item not in visitedNodes:
                visitedNodes.append(item)
                queue.append(item)
                
    return visitedNodes 

x = breadthFirstSearch(graphOfTree, '1')

print(x)
```
